### 2.2 Visualization the places of airports

```python
import folium
import pandas as pd

# Create a map centered at a specific location
world_map = folium.Map(location=[0, 0], zoom_start=2)

# Add markers for each airport
airport_data = pd.read_csv("airports.csv")

# airport_data = pd.read_csv("airports.csv")

# Create a map centered at a specific location
# Create a map centered at a specific location
world_map = folium.Map(location=[0, 0], zoom_start=2)

# Add markers for each airport
for index, airport in airport_data.iterrows():
    # Extract latitude and longitude from coordinates string
    coordinates = airport['coordinates'].split(',')
    latitude = float(coordinates[0].strip())
    longitude = float(coordinates[1].strip())

    color = 'green' if airport['continent'] == 'EU' else 'red'
    folium.CircleMarker(
        location=[latitude, longitude],
        radius=8,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=airport['iata_code'],
    ).add_to(world_map)

# Save the map to an HTML file or display it
world_map.save("airport_locations.html")
world_map.save("airport_locations.png")
```

### 2.3 change airport to airport_ID

```python
import pandas as pd

# Load the data from the 'new_output_with_sum_consum_customers.xlsx' file
file_path_airports = 'cleaned_airport.csv'
df_airports = pd.read_csv(file_path_airports)

# Create a dictionary mapping 'airport' to 'Airport_ID'
airport_mapping = dict(zip(df_airports['iata_code'], df_airports['Airport_ID']))

# Display the airport mapping
print("Airport Mapping Dictionary:")
print(airport_mapping)

# Load the data from the 'Passenger_EU.xlsx' file
file_path_passenger = 'new_passengersEU.csv'
df_passenger = pd.read_csv(file_path_passenger)

# Map 'airport' column to numerical values based on the created mapping
df_passenger['shopped_at'] = df_passenger['shopped_at'].map(airport_mapping)
df_passenger['departure_IATA_1'] = df_passenger['departure_IATA_1'].map(airport_mapping)
df_passenger['destination_IATA_1'] = df_passenger['destination_IATA_1'].map(airport_mapping)
df_passenger['destination_IATA_2'] = df_passenger['destination_IATA_2'].map(airport_mapping)

# Save the updated DataFrame to a new Excel file
updated_file_path_passenger = 'cleaned_Passenger_EU.xlsx'
df_passenger.to_excel(updated_file_path_passenger, index=False)

# Display the updated DataFrame
print("\nUpdated DataFrame with numerical values:")
print(df_passenger)
print(f'The new data has been saved to {updated_file_path_passenger}')
```
